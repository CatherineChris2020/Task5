**Category vs Sales Bar Graph**

```
plt.bar('Category','Sales',data=samp_data)
plt.title('Sales vs Category',size= 15)
plt.xlabel('Category',size= 13)
plt.ylabel('Sales',size= 13)
```

**Quantities Ordered Region Wise**

```
df_region=samp_data.groupby('Region')['Quantity'].sum().reset_index()
print(df_region)

labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['Quantity'],autopct='%1.1f%%',explode=(0.02,0.02,0.02,0.02),)
plt.title('Quantities ordered by each region',size=13)
```

**State wise Sales, Profit and Discount analysis**

```
s_data= samp_data.groupby(['State'])[['Sales', 'Discount','Profit']].mean()
```

**Displaying**

```
s_data
```

```
plt.figure(figsize = (20,14))
sns.violinplot(data =samp_data, x='State', y= 'Sales',palette='muted')
#plt.ylim(-1000,2300)
plt.xticks(rotation = 90)
plt.ylabel('Average Sales in Dollars')
plt.xlabel('States')
plt.show()
```

**Highest Selling Categories**

```
df_cats= samp_data.groupby('Category')['Quantity'].sum().reset_index()
df_cats
```

**Most Profitable Products**

```
dftop10_items = samp_data.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop10_items.reset_index(drop=True,inplace=True)
dftop10_items=dftop10_items.head(10)
dftop10_items
```

```
sns.pairplot(samp_data, hue="Ship Mode")
```

**Correlation**

```
samp_data.corr()
```

**Locationwise Analysis**

```
plt.figure(figsize = (16, 8))
sns.countplot(x = samp_data['State'],palette = 'rocket', order = (samp_data['State'].value_counts().head(50)).index)
plt.xticks(rotation=90,ha='right',fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('State',fontsize=15)
plt.ylabel('Quantities ordered',fontsize=15)
```
